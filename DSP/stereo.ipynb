```python
wavfile.write('channel_1.wav', s, ch1)
wavfile.write('channel_2.wav', s, ch2)
```

```python
ch1.dtype = np.int8
ch2.dtype = np.int8
```

```python
#Normalization of 8 bit data

ch1 = ch1 - 125
ch2 = ch2 - 125
```

```python
plt.close()
plt.plot(ch1, label="Light Channel")
plt.plot(ch2, label="Left channel")
plt.legend()
```

```python
plt.close()
ch1_fft = np.fft.fft(ch1)/len(ch1)
ffx = np.fft.fftfreq(len(ch1), 1/s)
plt.plot(ffx, np.abs(ch1_fft))
```

```python
plt.close()
ch2_fft = np.fft.fft(ch2)/len(ch2)
ffx = np.fft.fftfreq(len(ch2), 1/s)
plt.plot(ffx, np.abs(ch2_fft))
```
