```python
import numpy as np
import random
from scipy import signal
import scipy
import matplotlib.pyplot as plt
```

```python
gaussian_random_signal = []
for i in range(10000):
    gaussian_random_signal.append(random.gauss(0, 2))
```

```python
plt.plot(gaussian_random_signal)
```

```python
f, gauss_psd = signal.welch(gaussian_random_signal, 1/10, scaling="density")
plt.plot(f, gauss_psd)
```

```python
_ = plt.hist(gaussian_random_signal, bins=60)
```

```python
gauss_correlation = signal.correlate(gaussian_random_signal, gaussian_random_signal)
lags = signal.correlation_lags(len(gaussian_random_signal), len(gaussian_random_signal))
plt.plot(lags[9000:11000], gauss_correlation[9000:11000])
```
